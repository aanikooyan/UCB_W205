## Major questions in this project

### **Question 1: What are the 5 most popular trips that you would call "commuter trips"?**

To answer this question, the previously saved temporary table on the commuter trips (*commuter_trips_plus_distance*) is used (see Answer to _Q3-Part 2_ for more details).  Multiple queries across sequential steps are used to narrow down the problem and answer the question.

> Preview of the *commuter_trips_plus_distance* table

```python
%%bigquery

#standardSQL
SELECT
  trips.*,
  dist.distance
FROM
  `bikeshare_static.commuter_trips` AS trips
LEFT JOIN
  `bikeshare_static.station_dist` AS dist
ON
  trips.trip_id = dist.trip_id
ORDER BY
  distance DESC
LIMIT 5
```

> **Step 1**: Count all the trips between any pair of stations categorized based on the time of the trip (i.e. Morning or Afternoon trip)

```python
%%bigquery

#standardSQL
SELECT
  COUNT(trip_id) AS number_trips,
  start_station,
  end_station,
  trip_time
FROM (
  SELECT
    trip_id,
    corr_start_station_name AS start_station,
    corr_end_station_name AS end_station,
    start_hour_str,
    CASE
      WHEN start_hour_str LIKE '%Morning%' THEN 'Morning trip'
      WHEN start_hour_str LIKE '%Afternoon%' THEN 'Afternoon trip'
  END
    AS trip_time
  FROM
    `bikeshare_static.commuter_trips_plus_distance`)
GROUP BY
  start_station,
  end_station,
  trip_time
ORDER BY
  number_trips DESC
```

> **Step 2**: save the results from the query in Step 1 as a temporary table called *top_commuter_trips_step1*. Now, narrowing down the results even more by selecting top 20 most frequently used station pairs in each category of the _Morning trips_ and _Afternoon trips_.
>> NOTE: although the choice of top 20 station pairs for further analysis may seem arbitrary, my analysis with larger number of station pairs (50, 100, etc.) resulted the same at the end.

```python
%%bigquery df

# A query to select top 20 most frequently used station pairs in each category of the Morning trips and Afternoon trips
#standardSQL
SELECT
  *
FROM (
  SELECT
    *
  FROM
    `bikeshare_static.top_commuter_trips_step1`
  WHERE
    trip_time = 'Morning trip'
  ORDER BY
    number_trips DESC
  LIMIT
    20)
UNION ALL (
  SELECT
    *
  FROM
    `bikeshare_static.top_commuter_trips_step1`
  WHERE
    trip_time = 'Afternoon trip'
  ORDER BY
    number_trips DESC
  LIMIT
    20)
ORDER BY
  trip_time DESC
```

> **Step 3**: plot of the results from the query in Step 2 can help to get a rough idea of the reversed station pairs are related in the _Morning trips_ and _Afternoon trips_ categories.

```python
plt.scatter(data = df, x = 'start_station', y = 'end_station')
plt.xticks(rotation = 90)
plt.xlabel('Start stations', fontsize = 12, fontweight = 'bold')
plt.ylabel('End stations', fontsize = 12, fontweight = 'bold')
plt.grid()
```

> **Step 4**: We are so close to find the most popular commuter trips. To make it clear, the most popular commuter trips are defined as the ones with largest sum of the trips happening between the same pair of stations in the morning and in the afternoon throughout the entire time of data availability (i.e. 2013-2016). In other words, the total number of commuter trips between station A to B in the morning are summed with the ones from station B to A in the afternoon and will be sorted in descending order; finally, the top five station pairs will be selected as the 5 most popular commuter trips.

```python
df.info()
```

```python
# splitting the results into the morning and afternoon trips
df_am = df[df.trip_time == 'Morning trip']
df_pm = df[df.trip_time == 'Afternoon trip']
```

```python
'''
Finding all reversed pairs of stations in the morning and afternoon trips, 
and add their number of trips to each other and save them in a dictionary 
'''
d = dict()
count = 0
for index1, i in df_am.iterrows():
    for index2, j in df_pm.iterrows():
        if i[1]==j[2] and i[2]==j[1]:
            am = i[0]
            pm = j[0]
            d[(i[1], i[2])] = am+pm
print(d)
```

```python
# Create a DataFrame from the dictionary
top = pd.DataFrame.from_dict(d, orient = 'index', columns = ['total_trips'])
top.index.name = 'Pair of start/end stations'
```

```python
# Sort the results in descending order and pick the top 5 as the most popular commuter trips
top_5_commuter_trips = top.sort_values(by = 'total_trips', ascending = False).head(5)
top_5_commuter_trips
```

```python
# Plotting the total number of daily trips for top 5 popuar commuter trips
sns.barplot(x ='total_trips', y = 'Pair of start/end stations',data = top_5_commuter_trips.reset_index())
plt.xlabel('Total commuter trips', fontsize = 12, fontweight = 'bold')
plt.ylabel('Station pairs', fontsize = 12, fontweight = 'bold')
```

#### **Recommendation 1: _20-minute trips offer_**

Although most of the current offers have focused on the 30-minute basis, the analysis results showed that the majority of all trips (~ 91.3%) have taken less than or equal to 20 minutes. Thus, the company may consider modifying their offers or add new offers for 20-minute trips. This would include either commuter-type trips such as _Monthly/Annual Membership_ or short-term/daily type trips such as _Single Ride_ and _Access Pass_.

```python
%%bigquery
# A query to rank the trips based on trip time intervals
#standardSQL
SELECT
  trip_time_interval,
  trips,
  total_trips,
  ROUND(100 * trips/total_trips, 2) AS trips_percent
FROM (
  SELECT
    trip_time_interval,
    COUNT(*) AS trips
  FROM (
    SELECT
      CASE
        WHEN duration_sec <= 20 * 60 THEN ' t < 20 minutes'
        WHEN duration_sec <= 30 * 60 THEN '20 < t < 30 minutes'
        WHEN duration_sec <= 45 * 60 THEN '30 < t < 45 minutes'
        WHEN duration_sec <= 60 * 60 THEN '45 < t < 60 minutes'
      ELSE
      't > 60 minutes'
    END
      AS trip_time_interval
    FROM
      `bigquery-public-data.san_francisco.bikeshare_trips`)
  GROUP BY
    trip_time_interval),
  (
  SELECT
    COUNT(*) AS total_trips
  FROM
    `bigquery-public-data.san_francisco.bikeshare_trips`)
ORDER BY
  trips_percent DESC
```

```python
%%bigquery trip_per_month

# A query to find total number of bike trips per each month (of the year) across all time

#standardSQL
SELECT
  COUNT(*) AS num_trips,
  EXTRACT(MONTH
  FROM
    start_date) AS mon,
  FORMAT_DATETIME("%B",
    DATETIME(start_date)) AS month_name
FROM
  `bigquery-public-data.san_francisco.bikeshare_trips`
GROUP BY
  month_name,
  mon
ORDER BY
  mon
```

#### **Recommendation 3: _areas to focus_**

Many retails take advantage from the concept of _loyal customers_ or  _consistent users_ to increase their profitability. Such paradigm can also be recruited by *Lyft Bay Wheels* to come up with offers that focus more on the _consistent customers_. The result of my analysis showed that the zip code of the topmost consistent users within the commuter category perfectly matches the ones for non-commuters. This means that even _consistent_ non-commuters would most likely come from the same areas of the _loyal_ commuters. This finding can help the company to focus more on attracting customers living in those specific areas (zip codes). Subscribers in these areas may also specifically be considered with priority for the _Bike Share for All_ offer.

```python
%%bigquery

# Create a non-commuter trips table

#standardSQL
SELECT
  *
FROM (
  SELECT
    all_trips.*
  FROM
    `bikeshare_static.bikeshare_trips_corrected` AS all_trips
  LEFT JOIN
    `bikeshare_static.commuter_trips_plus_distance` AS commuter_trips
  ON
    all_trips.trip_id = commuter_trips.trip_id
  WHERE
    commuter_trips.trip_id IS NULL)

# Save results into the temporary table: non_commuter_trips 
```

```python
%%bigquery

# A query to find the (valid) subscriber zip code for the 5 topmosmt commuter vs non-commuter bikers

#standardSQL
SELECT
  commuter_trips.*,
  non_commuter_trips.*
FROM (
  SELECT
    COUNT(trip_id) AS num_trips_commuter,
    zip_code AS zip_commuter
  FROM
    `bikeshare_static.commuter_trips_plus_distance`
  WHERE
    -- to filter zip code entries with more or less than 5 characters
    zip_code LIKE '_____'
    -- to filter zip code entries with letters
    AND NOT REGEXP_CONTAINS(zip_code, r"[a-zA-Z]")
  GROUP BY
    zip_code
  ORDER BY
    num_trips_commuter DESC
  LIMIT
    5) AS commuter_trips
FULL OUTER JOIN (
  SELECT
    COUNT(trip_id) AS num_trips_non_commuter,
    zip_code AS zip_non_commuter
  FROM
    `bikeshare_static.non_commuter_trips`
  WHERE
    -- to filter zip code entries with more or less than 5 characters
    zip_code LIKE '_____'
    -- to filter zip code entries with letters
    AND NOT REGEXP_CONTAINS(zip_code, r"[a-zA-Z]")
  GROUP BY
    zip_code
  ORDER BY
    num_trips_non_commuter DESC
  LIMIT
    5) AS non_commuter_trips
ON
  commuter_trips.zip_commuter = non_commuter_trips.zip_non_commuter
ORDER BY
  num_trips_non_commuter DESC
```

#### **Recommendation 4: _focus simultaneously on both commuter and non-commuter trips_**:

Based on the results, although most of all trips are commuters (~ 61%), still the share of non-commuter trips is considerable. Thus, the recommendation here is to simultaneously focus on developing offers that would maximize the benefits from each of these types of bike trip. 

For example comparing the number of non-commuter trips in weekdays vs weekend shows that about 30% of all non-commuter trips happened during the weekend. Thus, the company may consider different _Single Ride_ or _Access Pass_ offers during the weekdays vs weekends to attract more customers in the weekends. 

```python
%%bigquery num_trips
# A query to compare the number of commuter trips vs non-commuter trips

#standardSQL
SELECT
  all_trips.*,
  commuter_trips.*
FROM (
  SELECT
    COUNT(*) AS number_all_trips
  FROM
    `bigquery-public-data.san_francisco.bikeshare_trips`) AS all_trips
FULL JOIN (
  SELECT
    COUNT(*) AS number_commuter_trips
  FROM
    `bikeshare_static.commuter_trips_plus_distance`) AS commuter_trips
ON
  number_all_trips = number_commuter_trips
```

```python
%%bigquery nct_wk

# A query to find the distribution of non-commuter trips between the weekdays and weekends

#standardSQL
SELECT
  COUNT(*) AS num_trips,
  day_type
FROM (
  SELECT
    trip_id,
    CASE
      WHEN EXTRACT(DAYOFWEEK FROM start_date) IN (1, 7) THEN "Weekend"
    ELSE
    "Weekday"
  END
    AS day_type,
  FROM
    `bikeshare_static.non_commuter_trips`)
GROUP BY
  day_type
```

#### **Recommendation 5: _keep the balance in the use of individual bikes_**:

As results show, the total number of times and/or hours each bike (out of 700 total bikes) has been used during the studied time period has almost a bimodal distribution, meaning that there is big gap between underusing and overusing bikes. Considering that the repair and maintenance of bikes can cost the company a lot, here I recommend the company to consider rotating the bikes among the most and least used stations on a regular basis. 

```python
%%bigquery bikes_frequency

# A query to rank individual bikes based on total number as well as total hour of use

#standardSQL
SELECT
  bike_number,
  COUNT(bike_number) AS total_number_used,
  CAST(ROUND(SUM(duration_sec)/3600) AS int64) AS total_hours_used
FROM
  `bigquery-public-data.san_francisco.bikeshare_trips`
GROUP BY
  bike_number
ORDER BY
  total_hours_used DESC,
  bike_number
```

#### **Recommendation 7: _keep the balance between bikes and docks availability at each station_**

From the analysis result, it has also been appeared that on average a good balance exists between the number of bikes and free docks available at each station. The sample distribution of the average percentages of bikes available at each station with respect to the total docks in that station show a close to normal distribution with mean and median of about 46% and 47%, meaning that most stations on average have kept the balance between the number of bikes available to rent and the number of free docks available to park the bike. The recommendation here would be use the dynamic data stream and time series data in order to keep this balance, specifically across the busy stations.

```python
%%bigquery bike_percent

#standardSQL
SELECT
  name,
  CAST(ROUND(AVG( bikes_available )) AS int64) AS mean_bikes_available,
  CAST(ROUND(AVG( docks_available)) AS int64) AS mean_docks_available,
  CAST(ROUND(AVG( total_docks)) AS int64) AS mean_total_docks,
  CAST(ROUND(AVG( percent_bikes_to_docks)) AS int64) AS mean_percent
FROM (
  SELECT
    bike_status_update.*
  FROM (
    SELECT
      bike_status.station_id,
      bike_stations.name,
      bike_status.bikes_available,
      bike_status.docks_available,
      bike_status.total_docks,
      bike_status.percent_bikes_to_docks
    FROM (
      SELECT
        station_id,
        bikes_available,
        docks_available,
        bikes_available + docks_available AS total_docks,
        CAST(ROUND(100* bikes_available/(bikes_available + docks_available)) AS int64) AS percent_bikes_to_docks,
      FROM
        `bigquery-public-data.san_francisco.bikeshare_status`
      WHERE
        docks_available <> 0) AS bike_status
    LEFT JOIN (
      SELECT
        station_id,
        name
      FROM
        `bigquery-public-data.san_francisco.bikeshare_stations`) AS bike_stations
    ON
      bike_status.station_id = bike_stations.station_id) AS bike_status_update)
GROUP BY
  name
ORDER BY
  mean_percent DESC
```
